## 通过爬虫获取数据

```python
import pandas as pd
import numpy as np
import re
import urllib
import bs4
```

```python
data_final = pd.DataFrame()
for year in range(2000,2022):
    for month in ['06','11']:
        url='https://www.top500.org/lists/top500/'+str(year)+'/'+month+'/' #访问从2000/06到2021/11的榜单
        response=urllib.request.urlopen(url)
        html=response.read()
        webpage=bs4.BeautifulSoup(html,'html.parser')


        #爬取第一部分数据
        table = webpage.find('table', class_= "table table-condensed table-striped") #在页面中获取需要的表格部分
        col = list(filter(None,table.contents[1].text.split('\n')))
        col.insert(0,'Year')
        col.insert(1,'Month')
        data = pd.DataFrame(columns = col)
        data = data.rename(columns={'System':'Country'})
        for i in range(3,len(table.contents),2):
            tmptd = table.contents[i].find_all('td')
            if len(tmptd) != 6:
                print('\npart1 error!!!!!!!!') #如果爬虫中某一页结构发生变化，即print异常，并且跳过此次循环
                print(year,month,i) 
                continue
            tmp = [tmptd[j].text for j in range(len(tmptd))]

            #对于表中第一列只提取国家信息
            tmp1 = tmptd[1].find_all('a')[1].next_sibling.next_sibling.split('\n')[0]
            tmp[1] = tmp1

            tmp.insert(0,year)
            tmp.insert(1,int(month))
            tmpdict = dict(zip(data.columns,tmp))
            data = data.append(tmpdict,ignore_index=True)


        #爬取第二部分数据，即更详细的数据['Rank','Manufacturer','OperatingSystem','Memory','Processor']
        data2 = pd.DataFrame()
        list1 = ['Manufacturer','Operating System','Memory','Processor'] #只想爬取这部分数据
        for i in range(3,len(table.contents),2):
            tmpurl = table.contents[i].find_all('a')[0]['href']
            tmpurl = 'https://www.top500.org/'+tmpurl #即进入每一行中更细节的页面
            response=urllib.request.urlopen(tmpurl)
            tmphtml=response.read()
            tmpwebpage=bs4.BeautifulSoup(tmphtml,'html.parser')
            tmptable = tmpwebpage.find('table',class_ = 'table table-condensed') #因为验证过只会找到一个，所以只用find
            name = ['Year','Month']
            val = [year,int(month)]
            name.append('Rank')
            val.append(table.contents[i].find('td').text)
            name = name + list1
            for item in list1:
                tmpth = tmptable.find('th',text = re.compile(item))
                if tmpth == None:
                    print('\npart2 error!!!!!!!')
                    print(year,month,i,item)
                    val.append(np.nan)
                else:
                    tmptd = tmpth.next_sibling.next_sibling
                    clean_pattern = re.compile(r'\n| |\xa0|\\xa0|\u3000|\\u3000|\\u0020|\u0020|\t|\r') #构建一个包含各种比如换行 空格的pattern
                    clean_tmptd = clean_pattern.sub('', tmptd.text)  #利用sub函数 表明将tmptr[j].text中所有匹配上clean_pattern的都替换成‘’
                    val.append(clean_tmptd)
            if len(name)!=len(val):
                raise
            tmpdict = dict(zip(name,val))
            data2 = data2.append(tmpdict,ignore_index=True)
    
        print(data.shape)
        print(data2.shape)
        #将data和data2合并
        data_finaltmp = pd.merge(data,data2,on = ['Year','Month','Rank'],how = 'inner')
        print(data_finaltmp.shape)
        data_final = data_final.append(data_finaltmp)
print(data_final.shape)
data_final.to_csv('./data/2_data_final.csv',index = False)
```

变量说明：
Rmax最大性能 Rpeak峰值性能  --衡量计算机算力
OperatingSystem 操作系统
Processor 处理器

```python
#对‘Memory’变量做处理
data_final2['Memory'] = data_final2['Memory'].apply(lambda x: x.split('GB')[0])
data_final2 = data_final2.rename(columns={'Memory':'Memory (GB)'})
#将数值型变量转为数值
for i in ['Cores','Power (kW)','Memory (GB)','Rmax (GFlop/s)','Rpeak (GFlop/s)','Rmax (TFlop/s)','Rpeak (TFlop/s)']:
    print(i)
    data_final2.loc[~data_final2[i].isnull(),i] = data_final2.loc[~data_final2[i].isnull(),i].apply(lambda x: x.replace(',','')) #去掉数值型字符中的','
    data_final2.loc[~data_final2[i].isnull(),i] = data_final2.loc[~data_final2[i].isnull(),i].astype(float)

##处理‘Rmax’‘Rpeak’两列的缺失情况
index1= data_final2[data_final2['Rmax (GFlop/s)'].isnull()].index
index2 = data_final2[data_final2['Rmax (TFlop/s)'].isnull()].index
data_final2.loc[index1,'Rmax (GFlop/s)']=data_final2.loc[index1,'Rmax (TFlop/s)']*1000
data_final2.loc[index1,'Rpeak (GFlop/s)']=data_final2.loc[index1,'Rpeak (TFlop/s)']*1000
data_final2.loc[index2,'Rmax (TFlop/s)']=data_final2.loc[index2,'Rmax (GFlop/s)']/1000
data_final2.loc[index2,'Rpeak (TFlop/s)']=data_final2.loc[index2,'Rpeak (GFlop/s)']/1000
```

```python
data_final2 = data_final2.drop(['Rmax (GFlop/s)','Rpeak (GFlop/s)'],1)
data_final2.head(3)
```

```python
data = data_final.copy()
data.head(3)
```

```python
#Top10榜单各年平均性能及增长率变化
groups = data.groupby('Year')
y1 = groups['Rmax (TFlop/s)'].mean()
y2 =  groups['Rpeak (TFlop/s)'].mean()
x = y1.index.tolist()

#计算增长率
y3 = y1.pct_change()
y4 = y2.pct_change()
y5 = (y3+y4)/2

y1 = y1.apply(lambda x:'%.4f' % x).apply(lambda x:float(x))
y2 = y2.apply(lambda x:'%.4f' % x).apply(lambda x:float(x))
y5 = y5.apply(lambda x:'%.4f' % x).apply(lambda x:float(x))

data_Rmean = pd.DataFrame(columns=['Year','Rmax (TFlop/s)','Rpeak (TFlop/s)','平均增长率'])
data_Rmean['Year'] = x
data_Rmean['Rmax (TFlop/s)'] = y1.values.tolist()
data_Rmean['Rpeak (TFlop/s)'] = y2.values.tolist()
data_Rmean['平均增长率'] = y5.values.tolist()
data_Rmean
data_Rmean.to_csv('./data/2_dataRmean.csv',index=False)
```

```python
#2000年至2021年各国在前10榜单的个数
countrylist = data[(data['Year'] == 2000) | (data['Year'] == 2021)]['Country'].unique().tolist()

y = data.groupby(['Year','Country']).size()
data_country = y.reset_index()
data_country = data_country.rename(columns={0:'count'})
data_country  = data_country[data_country['Year'].apply(lambda x : x in [2000,2005,2010,2015,2020])]
print(data_country['Country'].unique())
#主要想看 China，United States，Japan，将Germany，Netherlands（荷兰），Spain（西班牙），Switzerland（瑞士），France（法国），Saudi Arabia（沙特阿拉伯），Italy 欧亚其它国家
maplist = {'China':'中国','United States':'美国','Japan':'日本'}
data_country['Country'] = data_country['Country'].map(maplist).fillna('其它国家')
maplist = {'中国':2,'美国':1,'日本':3,'其它国家':4}
data_country['id'] = data_country['Country'].map(maplist)
data_country.to_csv('./data/2_data_country.csv',index = False)
```

```python
#2000年至2021年各个公司在前10榜单的个数
data['Manufacturer'].unique()
y = data.groupby(['Year','Manufacturer']).size()
data_fa = y.reset_index()
data_fa = data_fa.rename(columns={0:'count'})
data_fa  = data_fa[data_fa['Year'].apply(lambda x : x in [2000,2005,2010,2015,2020])]
print(data_fa['Manufacturer'].unique())
data_fa[data_fa['count']>3]
data_fa.to_csv('./data/2_datafa.csv',index = False)
```

```python
#2000年至2021年各个操作系统在前10榜单的个数
print(data['Operating System'].unique())
#进行分组处理
data['Operating System2'] = data['Operating System']
data['Operating System2'] = data['Operating System2'].apply(lambda x: 'SUSELinux' if 'SUSELinux' in x else x)
data['Operating System2'] = data['Operating System2'].apply(lambda x: 'RedHat' if ('RHEL' in x)|('RedHat' in x) else x)
data['Operating System2'] = data['Operating System2'].apply(lambda x: 'CentOS' if 'CentOS' in x else x)
data['Operating System2'] = data['Operating System2'].apply(lambda x: 'Ubuntu' if 'Ubuntu' in x else x)
print(data['Operating System2'].unique())

y = data.groupby(['Year','Operating System2']).size()
data_system = y.reset_index()
data_system = data_system.rename(columns={0:'count'})
data_system  = data_system[data_system['Year'].apply(lambda x : x in [2000,2005,2010,2015,2020])]
data_system.to_csv('./data/2_datasystem.csv',index = False)
```

```python
#2000年至2021年各个处理器在前10榜单的个数
print(data['Processor'].unique())
#进行分组处理
data['Processor2'] = data['Processor']
data['Processor2'] = data['Processor2'].apply(lambda x: 'POWERPC' if ('Power' in x)|('POWER' in x) else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'Hitachi' if 'Hitachi' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'Opteron' if 'Opteron' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'Xeon' if 'Xeon' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'Itanium' if 'Itanium' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'AMDEPYC' if 'AMDEPYC' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'Alpha' if 'Alpha' in x else x)
data['Processor2'] = data['Processor2'].apply(lambda x: 'SPARC' if 'SPARC' in x else x)
print(data['Processor2'].unique())

y = data.groupby(['Year','Processor2']).size()
data_processor = y.reset_index()
data_processor = data_processor.rename(columns={0:'count'})
data_processor  = data_processor[data_processor['Year'].apply(lambda x : x in [2000,2005,2010,2015,2020])]
data_processor.to_csv('./data/2_dataprocessor.csv',index = False)
```
